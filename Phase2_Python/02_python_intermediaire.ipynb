```python
# GÉNÉRATEURS
# Produisent des valeurs à la demande (lazy evaluation)
# Ne chargent pas tout en mémoire → parfait pour les grands datasets ML

# Fonction normale vs générateur
def liste_normale(n):
    result = []
    for i in range(n):
        result.append(i**2)
    return result   # crée tout en mémoire

def generateur(n):
    for i in range(n):
        yield i**2  # produit une valeur à la fois, pause, reprend

# Comparaison mémoire
import sys

liste = liste_normale(10000)
gen = generateur(10000)

print(f"Liste normale : {sys.getsizeof(liste):,} bytes")
print(f"Générateur    : {sys.getsizeof(gen):,} bytes")
print(f"→ Le générateur utilise ~{sys.getsizeof(liste)//sys.getsizeof(gen)}x moins de mémoire !")

print("---")

# Utiliser un générateur
gen = generateur(5)
print("Valeurs du générateur :")
for valeur in gen:
    print(f"  {valeur}")

print("---")

# Generator expression (comme list comprehension mais lazy)
gen_exp = (x**2 for x in range(1000000))  # ne crée rien en mémoire
print(f"Generator expression : {sys.getsizeof(gen_exp)} bytes")
print(f"Première valeur : {next(gen_exp)}")
print(f"Deuxième valeur : {next(gen_exp)}")

print("---")

# Générateur de batches — utilisé en deep learning !
def batch_generator(data, batch_size):
    for i in range(0, len(data), batch_size):
        yield data[i:i+batch_size]

dataset = list(range(100))
print("Batches de taille 10 :")
for i, batch in enumerate(batch_generator(dataset, 10)):
    print(f"  Batch {i} : {batch[:3]}... ({len(batch)} éléments)")
```

```python
# DÉCORATEURS
# Un décorateur = une fonction qui modifie le comportement d'une autre fonction

# Décorateur basique
def mon_decorateur(func):
    def wrapper(*args, **kwargs):
        print(f"→ Avant l'appel de {func.__name__}")
        result = func(*args, **kwargs)
        print(f"→ Après l'appel de {func.__name__}")
        return result
    return wrapper

@mon_decorateur
def saluer(nom):
    print(f"Bonjour, {nom} !")

saluer("Zooher")

print("---")

# Décorateur de timing — très utile pour mesurer la performance
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        elapsed = time.time() - start
        print(f"⏱ {func.__name__} a pris {elapsed:.4f} secondes")
        return result
    return wrapper

@timer
def calcul_lent():
    total = sum(range(1000000))
    return total

resultat = calcul_lent()
print(f"Résultat : {resultat}")

print("---")

# Décorateur de cache (mémoïsation)
from functools import lru_cache

@lru_cache(maxsize=128)
def fibonacci(n):
    if n < 2:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print("Fibonacci sans cache serait très lent pour n=35")
start = time.time()
print(f"fibonacci(35) = {fibonacci(35)}")
print(f"⏱ Temps : {time.time()-start:.6f} secondes")
```

```python
# CONTEXT MANAGERS
# Gestion automatique des ressources (fichiers, connexions, timers)
# Le "with" garantit que la ressource est libérée même si une erreur survient

# Context manager basique avec fichier
print("=== FICHIERS ===")
with open("test.txt", "w") as f:
    f.write("Hello ML !\n")
# fichier automatiquement fermé ici, même si erreur

with open("test.txt", "r") as f:
    print(f.read())

import os
os.remove("test.txt")

# Créer son propre context manager
from contextlib import contextmanager
import time

@contextmanager
def timer(nom):
    print(f"⏱ Début : {nom}")
    start = time.time()
    yield                           # code du bloc with s'exécute ici
    elapsed = time.time() - start
    print(f"✅ Fin : {nom} → {elapsed:.4f}s")

# Utilisation
with timer("Calcul numpy"):
    import numpy as np
    result = np.sum(np.random.randn(1000000))
    print(f"   Résultat : {result:.4f}")

with timer("Tri d'une liste"):
    data = list(range(100000, 0, -1))
    data.sort()
    print(f"   Premier élément trié : {data[0]}")

print("---")

# Context manager pour ML — sauvegarder/restaurer état
@contextmanager
def numpy_seed(seed):
    state = np.random.get_state()   # sauvegarder l'état
    np.random.seed(seed)
    yield
    np.random.set_state(state)      # restaurer l'état

with numpy_seed(42):
    print("Avec seed 42 :", np.random.randn(3))

with numpy_seed(42):
    print("Même seed 42 :", np.random.randn(3))

print("Résultats identiques → reproductibilité garantie !")
```

```python
# MODULES ET PACKAGES
# Un module = un fichier Python réutilisable
# Un package = un dossier de modules

import os

# Créer notre propre module utilitaire
module_code = '''
"""
Module utilitaire pour le ML
"""
import numpy as np

def normaliser(X):
    """Normalise les données entre 0 et 1"""
    return (X - X.min()) / (X.max() - X.min())

def standardiser(X):
    """Standardise les données (μ=0, σ=1)"""
    return (X - X.mean()) / X.std()

def train_test_split_simple(X, y, test_size=0.2, seed=42):
    """Split simple sans scikit-learn"""
    np.random.seed(seed)
    n = len(X)
    indices = np.random.permutation(n)
    split = int(n * (1 - test_size))
    train_idx = indices[:split]
    test_idx = indices[split:]
    return X[train_idx], X[test_idx], y[train_idx], y[test_idx]

def accuracy(y_true, y_pred):
    """Calcule l'accuracy"""
    return np.mean(y_true == y_pred)

def afficher_stats(X, nom="Dataset"):
    """Affiche les statistiques d\'un tableau"""
    print(f"=== {nom} ===")
    print(f"Shape  : {X.shape}")
    print(f"Mean   : {X.mean():.4f}")
    print(f"Std    : {X.std():.4f}")
    print(f"Min    : {X.min():.4f}")
    print(f"Max    : {X.max():.4f}")

ML_TOOLS_VERSION = "1.0.0"
'''

# Sauvegarder le module
with open("ml_tools.py", "w") as f:
    f.write(module_code)

print("✅ Module ml_tools.py créé !")

# Utiliser notre module
import importlib.util
spec = importlib.util.spec_from_file_location("ml_tools", "ml_tools.py")
ml_tools = importlib.util.module_from_spec(spec)
spec.loader.exec_module(ml_tools)

# Tester les fonctions
import numpy as np
X = np.random.randn(100, 3)
y = (X[:, 0] > 0).astype(int)

print(f"\nVersion : {ml_tools.ML_TOOLS_VERSION}")
ml_tools.afficher_stats(X, "Dataset brut")

X_norm = ml_tools.normaliser(X)
ml_tools.afficher_stats(X_norm, "Dataset normalisé")

X_std = ml_tools.standardiser(X)
ml_tools.afficher_stats(X_std, "Dataset standardisé")

X_train, X_test, y_train, y_test = ml_tools.train_test_split_simple(X, y)
print(f"\nSplit : Train={len(X_train)} | Test={len(X_test)}")

# Packages — structure d'un vrai projet ML
print("\n=== STRUCTURE D'UN VRAI PROJET ML ===")
print("""
mon_projet_ml/
├── data/
│   ├── raw/          ← données brutes
│   └── processed/    ← données préprocessées
├── notebooks/        ← Jupyter notebooks d'exploration
├── src/
│   ├── __init__.py
│   ├── data.py       ← chargement des données
│   ├── features.py   ← feature engineering
│   ├── models.py     ← définition des modèles
│   └── evaluate.py   ← métriques d'évaluation
├── models/           ← modèles sauvegardés
├── tests/            ← tests unitaires
├── requirements.txt  ← dépendances
└── README.md         ← documentation
""")
```
