```python
# PCA — Principal Component Analysis
# Réduire des données de haute dimension en gardant l'essentiel

# Simuler un dataset : 100 exemples, 3 features
np.random.seed(42)
X = np.random.randn(100, 3)

# Étape 1 : centrer les données (soustraire la moyenne)
X_centre = X - np.mean(X, axis=0)

# Étape 2 : calculer la matrice de covariance
cov_matrix = np.cov(X_centre.T)
print("Matrice de covariance :\n", cov_matrix)

# Étape 3 : valeurs et vecteurs propres de la covariance
valeurs, vecteurs = np.linalg.eig(cov_matrix)
print("\nValeurs propres :", valeurs)

# Étape 4 : variance expliquée par chaque composante
variance_expliquee = valeurs / np.sum(valeurs)
print("\nVariance expliquée :", variance_expliquee)
print("Variance cumulée :", np.cumsum(variance_expliquee))

# Étape 5 : projeter sur les 2 premières composantes
X_pca = X_centre @ vecteurs[:, :2]
print("\nShape original :", X.shape)
print("Shape après PCA :", X_pca.shape)
```

```python
# Valeurs propres et vecteurs propres
# Pour une matrice A : A·v = λ·v
# v = vecteur propre (ne change pas de direction)
# λ = valeur propre (facteur d'échelle)

A = np.array([[4, 2],
              [1, 3]])

valeurs_propres, vecteurs_propres = np.linalg.eig(A)

print("Matrice A :\n", A)
print("\nValeurs propres (λ) :", valeurs_propres)
print("\nVecteurs propres (v) :\n", vecteurs_propres)

# Vérification : A·v = λ·v
v0 = vecteurs_propres[:, 0]   # premier vecteur propre
λ0 = valeurs_propres[0]       # première valeur propre

print("\nVérification A·v0 :", A @ v0)
print("λ0·v0 :", λ0 * v0)
```

```python
import numpy as np

# Un vecteur = une liste de nombres
# En ML, chaque exemple de données est représenté comme un vecteur

v=np.array([1,2,3])
print("Vecteur :", v)
print("Type :", type(v))
print("Forme (shape) :", v.shape)
print("Dimension :", v.ndim)
```

```python
#Les Operations sur les vecteurs
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("a :", a)
print("b :", b)
print("a + b :", a + b)
print("a * 2 :", a * 2)
print("a * b :", a * b)

# Produit scalaire (dot product) - tres utilisé en ML
dot = np.dot(a, b)
print("Prodiot scalaire a.b :", dot)

# Norme (longueur du vecteur)
norme = np.linalg.norm(a)
print("Norme de a :", norme)
```

```python
# Une matrice = tableau de nombres à 2 dimensions
# En ML : un dataset entier est une matrice (lignes=exemples, colonnes=features)

M = np.array([[1, 2, 3],
              [4, 5, 6]])

print("Matrice M :")
print(M)
print("Shape :", M.shape)   # (2 lignes, 3 colonnes)
print("Dimensions :", M.ndim)

# Accéder aux éléments
print("Ligne 0 :", M[0])
print("Élément ligne 1, colonne 2 :", M[1, 2])
print("Transposée :\n", M.T)
```

```python
# Multiplication matricielle
# Règle : A (m×n) @ B (n×p) = C (m×p)
# Les colonnes de A doivent égaler les lignes de B

A = np.array([[1, 2],
              [3, 4],
              [5, 6]])   # shape (3, 2)

B = np.array([[7, 8, 9],
              [10, 11, 12]])  # shape (2, 3)

C = A @ B   # shape résultat : (3, 3)

print("A shape:", A.shape)
print("B shape:", B.shape)
print("A @ B =")
print(C)
print("C shape:", C.shape)
```
